## Imports

````python
# import packages
import numpy as np
````

````python
d1 = np.zeros((10,100)) # a 10 x 100 matrix
d2 = np.zeros(100) # an array of size 100

# vectorized operations are your best friend (for time savings)
d3 = d2 + 1 # an array of 100 ones
d4 = d2 + d3 # an array of 100 ones
d5 = 36 * d4 # an array of 100 thirty-sixes
d6 = np.sqrt(d5) # an array of 100 sixes
d7 = d6 > 5 # an array of 100 Trues
````

````python
flora_mileage = [0.5, 0.7, 0.2, 0.9, 0.3] # a list

# a for loop
for i, miles in enumerate(flora_mileage): # enumerate simplifies indexing (returns a tuple (index, value))
	if miles > 0.5:
		# f-strings simplify printing non-strings with strings
		print(f"on day {i+1}, Flora walked a lot ({miles} miles)")
	elif miles > 0.2:
		print(f"on day {i+1}, Flora walked a moderate amount ({miles} miles)")
	else:
		print(f"on day {i+1}, Flora walked little ({miles} miles)")

# list comprehensions simplify for-loops
flora_mileage_large = [miles for miles in flora_mileage if miles > 0.5]
print(f"flora_mileage_large: {flora_mileage_large}")
````

## Functions + Assertions

````python
# barebones function syntax
def sum(a, b):
    return a + b

def sum_error(a, b):
    return a * b

# assertions help you catch logical errors early 
assert sum(3, 4) == 7 # no output indicates no error
assert sum_error(3, 4) == 7, "the sum is false" # raises AssertionError
````
